#### Splitting train and test data

```python
y = bos['Price ($1000s)']
```

```python
#X = bos.drop(["RAD","DIS","NOX","Price ($1000s)"],axis=1)
```

```python
X = bos.drop(['Price ($1000s)'],axis=1)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=1)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
## fitting the LR model
regression = LinearRegression()
regression.fit(X_train,y_train)
```

```python
## saving the model in local file system
filename = 'linear_regression.pickle'
pickle.dump(regression,open(filename,'wb'))
```

```python
## prediction using saved model
#loaded_model = pickle.load(open(filename,'rb'))
#y_predict = loaded_model.predict(X_test)
#y_predict
```

```python
## prediction using saved model
loaded_model = pickle.load(open(filename,'rb'))
y_predict = loaded_model.predict(X_test)
y_predict
```

```python
def adj_r2(x,y):
    r2 = regression.score(x,y)
    n = x.shape[0]
    p = x.shape[1]
    adjusted_r2 = 1-(1-r2)*(n-1)/(n-p-1)
    return adjusted_r2
```

```python
## r2 and adj r2 with dropping 3 features
```

```python
## For training data
regression.score(X_train,y_train)
```

```python
adj_r2(X_train,y_train)
```

```python
## For test data
regression.score(X_test,y_test)
```

```python
adj_r2(X_test,y_test)
```

```python
## r2 and adj r2 without dropping 3 features
```

```python
## For training data
regression.score(X_train,y_train)
```

```python
adj_r2(X_train,y_train)
```

```python
## for test data
regression.score(X_test,y_test)
```

```python
adj_r2(X_test,y_test)
```

#### It seems that there is a better results in r2 and adj r2 without removing the features.
